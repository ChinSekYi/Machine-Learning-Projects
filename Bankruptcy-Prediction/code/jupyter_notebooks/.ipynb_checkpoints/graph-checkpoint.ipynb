```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram

df.iloc[:, 0:10].hist(figsize=(10, 6))  # figsize is optional, adjust as needed
plt.show()
```

```python
# remove outlier
import pandas as pd
import matplotlib.pyplot as plt


# Define a function to remove outliers based on 1.5 * IQR for each feature
def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column >= lower_bound) & (column <= upper_bound)]


# Remove outliers for each feature and plot histograms
clean_list = []
for column in df.columns:
    cleaned_data = remove_outliers(df[column])
    clean_list.append(cleaned_data)

plt.figure(figsize=(15, 40))
for i, series in enumerate(clean_list, start=1):
    plt.subplot(22, 3, i)  # Adjust layout as needed
    plt.hist(series, bins=10, alpha=0.5)
    plt.title(f"Histogram of {series.name}", fontsize=8)
    plt.xlabel("Value")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 40))
for i, series in enumerate(clean_list, start=1):
    plt.subplot(22, 3, i)  # Adjust layout as needed
    sns.kdeplot(series, color="skyblue")
    plt.title(f"Density plot of {series.name}", fontsize=8)
    plt.xlabel("Value")
    plt.ylabel("Frequency")


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 40))
for i, series in enumerate(clean_list, start=1):
    plt.subplot(22, 3, i)  # Adjust layout as needed
    sns.boxplot(data=series, color="skyblue")
    plt.title(f"Box Plot of {series.name}", fontsize=8)
    plt.ylabel("Value")

plt.tight_layout()
plt.show()
```

```python
# ECDF
def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(data) + 1) / len(data)
    return x, y


plt.figure(figsize=(15, 40))
for i, data in enumerate(clean_list, start=1):
    plt.subplot(22, 3, i)  # Adjust layout as needed
    x, y = ecdf(data)
    plt.plot(x, y, marker=".", linestyle="none", linewidth=0.2)
    plt.xlabel("Value")
    plt.ylabel("ECDF")
    plt.title(f"ECDF of {series.name}")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# remove outlier (for the whole df)
import pandas as pd
import matplotlib.pyplot as plt

# Define a function to remove outliers based on 1.5 * IQR for each feature


def remove_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[((df >= lower_bound) & (df <= upper_bound)).all(axis=1)]


df_remove = remove_outliers(df)
len(df_remove)
```

## remove outliers analysis
After initially removing outliers from the dataset, we observed that a significant portion of the data was eliminated, resulting in a substantially reduced dataset size. This prompted a reconsideration of the outlier removal process, as it became apparent that the removed data points may have had meaningful contributions to the analysis. Given the potential significance of outliers in influencing results, we decided to retain the original dataset without removing outliers. This decision ensures that the analysis encompasses the full range of data points, allowing for a more comprehensive exploration of the underlying patterns and insights within the dataset.
